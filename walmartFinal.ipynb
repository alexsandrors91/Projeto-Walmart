É solicitado pela empresa walmart a análise das vendas semanais para um possivel investimento nas lojas em que as vendas são mais promissoras.



Primeiro é necessário importar as bibliotecas que serão utilizadas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

É necessário alterar alguns formatos de coluna do data set

```python
data['Date'] = data['Date'].astype('datetime64')
data['Holiday_Flag'] = data['Holiday_Flag'].astype('bool')
data['Weekly_Sales'] = data['Weekly_Sales'].astype('int')
```

```python
data.info()
```

Verificando preço médio de venda semanal por loja

```python
group_med_sales = data.groupby("Store").mean().reset_index()
group_med_sales
```

Qual o atributo que tem maior correlação (positiva ou negativa) com o número de vendas? (E que faça sentido)

```python
correlacao = data.corr(method='kendall', min_periods=1000)
display(correlacao['Weekly_Sales'])
```

A maior correlação é holiday flag

Da loja com maior venda acumulada (soma de vendas de todo o período), quantas semanas do ano ele ultrapassou a média do período:


```python
order = data.sort_values("Weekly_Sales", ascending = False)
group_order = data.groupby("Store").sum().reset_index()
group_order = data.sort_values("Weekly_Sales", ascending = False)
group_order = group_order.head(5)
group_order
```

Avaliando loja com maiores vendas semanais

```python
l14 = data[data['Store'] == 14]
med = data.groupby('Weekly_Sales').mean().reset_index()
media = med['Weekly_Sales']
l14['media'] = media
l14 = l14.sort_values('Holiday_Flag', ascending = False);
l14
```

Faça um gráfico com as vendas máximas, mínimas e médias de todas as lojas ao longo do tempo. Ressalte de alguma forma, o período de feriados mencionados acima.

```python
media_vendas = data.groupby("Store").mean().reset_index()
max_vendas = data.groupby("Store").sum().reset_index()
min_vendas = max_vendas.sort_values('Weekly_Sales', ascending = True)
max_vendas
```
